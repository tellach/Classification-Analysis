```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
customer_data = pd.read_csv('./shopping.csv')
```

```python
customer_data.shape
```

```python
customer_data.head()
```

```python
data = customer_data.iloc[:, 3:5].values
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward')) #inertie interclasse maximale methode de ward
```

```python
#Apres obtention de 5 clusters, on va faire un groupement de données pour ces 5 clusters
#Nous allon utiliser scikit-learn pour ça 
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
data.shape
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```
